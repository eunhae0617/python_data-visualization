```python
biz19=food2019.loc[:,['biz_mid_nm','use_amt','time']].groupby(['biz_mid_nm','time']).agg('mean').astype(int)

biz20=food2020.loc[:,['biz_mid_nm','use_amt','time']].groupby(['biz_mid_nm','time']).agg('mean').astype(int)
```

```python
biz19=pd.DataFrame(food2019)
biz20=pd.DataFrame(food2020)
biz19=food2019.reset_index()
biz20=food2020.reset_index()
```

```python
type=food2019['biz_mid_nm'].unique()
```

```python
fig, ax=plt.subplots(nrows=3,ncols=5)
c=0
r=0
for t in type:
    sns.lineplot(x="time",y="use_amt",data=biz19[biz19['biz_mid_nm']==t],ax=ax[r,c],legend='brief',label='2019')
    sns.lineplot(x="time",y="use_amt",data=biz20[biz20['biz_mid_nm']==t],ax=ax[r,c],legend='brief',label='2020')  
    ax[r,c].set_title(t)
    c=c+1
    if c==5:
        r=r+1
        c=0
    
plt.show()
```
